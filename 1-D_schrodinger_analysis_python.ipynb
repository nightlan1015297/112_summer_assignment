```python
import numpy as np
from scipy.sparse.linalg import eigsh
from scipy import sparse
import plotly.graph_objects as go
```

```python
N_points = 1000
up_bound = 1
dn_bound = -1

k = 1
m = 1

del_s = (up_bound - dn_bound)/(N_points - 1)

x_grid = np.linspace(dn_bound,up_bound,N_points,dtype=float)

D = sparse.spdiags(
    np.array([np.ones(N_points), -2*np.ones(N_points), np.ones(N_points)]),
    np.array([-1, 0,1]),
    N_points, N_points)

T_hat = (-1/2) * D /del_s**2

def potential (x_grid):
    return k*(x_grid)**2/2

V_hat = sparse.diags(potential(x_grid),(0))

H_hat = T_hat + V_hat

E , psi = eigsh(H_hat,which='SM')

```

```python
def Hermite(n,x):
    if n == 0:
        return 1
    elif n == 1:
        return x
    elif n == 2:
        return 4*x**2-2
omega = np.sqrt(k/m)
h_bar = 1

def analytic_sol(n,x):
    return 1/(np.sqrt(2**n*np.math.factorial(n))) * (m*omega/(np.pi*h_bar))**(1/4)*np.exp(-(m*omega*x**2)/(2*h_bar)) * Hermite(n,(np.sqrt(m*omega/h_bar))*x)
```

```python
#plot the wave function from the eigen function

fig = go.Figure()
fig.add_trace(go.Scatter(x=x_grid, y=psi[:,0], name="psi_0"))
fig.add_trace(go.Scatter(x=x_grid, y=psi[:,1], name="psi_0"))
fig.add_trace(go.Scatter(x=x_grid, y=psi[:,2], name="psi_0"))
fig.add_trace(go.Scatter(x=x_grid, y=analytic_sol(0,x_grid), name="psi_0"))
fig.update_layout(
    title="Wave function",
    xaxis_title="x",
    yaxis_title="psi",
    legend_title="Legend Title",
    font=dict(
        family="Courier New, monospace",
        size=18,
        color="RebeccaPurple"
    )
)
fig.show()
```
